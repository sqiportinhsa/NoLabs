# Сравнение сортировок
## Загрузка данных из файлов

```python
import matplotlib.pyplot as plt
import csv
```

```python
bubble_x = []
bubble_y = []
insertion_x = []
insertion_y = []
selection_x = []
selection_y = []


with open("../testing/data/bench11.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
        
        if (row[0] == "Bubble sort"):
            bubble_x.append(x)
            bubble_y.append(y)
        elif (row[0] == "Insertion sort"):
            insertion_x.append(x)
            insertion_y.append(y)
        elif (row[0] == "Selection sort"):
            selection_x.append(x)
            selection_y.append(y)
```

```python
ran_x = []
ran_y = []
cen_x = []
cen_y = []
med_x = []
med_y = []


with open("../testing/data/bench2.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
        
        if (row[0] == "Random pivot qsort"):
            ran_x.append(x)
            ran_y.append(y)
        elif (row[0] == "Central pivot qsort"):
            cen_x.append(x)
            cen_y.append(y)
        elif (row[0] == "Median pivot qsort"):
            med_x.append(x)
            med_y.append(y)
            
```

```python
merge_x = []
merge_y = []

with open("../testing/data/bench3.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
    
        merge_x.append(x)
        merge_y.append(y)
```

```python
lsd_x = []
lsd_y = []

with open("../testing/data/bench4.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
    
        lsd_x.append(x)
        lsd_y.append(y)
```

## Quadratic sortings

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('Quadratic sortings comparison', fontsize = 15)

plt.plot(bubble_x, bubble_y)
plt.plot(insertion_x, insertion_y)
plt.plot(selection_x, selection_y)

plt.legend(['bubble', 'insertion', 'selection'])

plt.show()
```

## QSORT

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('Qsort sortings comparison', fontsize = 15)

plt.plot(ran_x, ran_y)
plt.plot(cen_x, cen_y)
plt.plot(med_x, med_y)

plt.legend(['random', 'central', 'median'])

plt.show()
```

## MergeSort


```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('Merge sort plot', fontsize = 15)

plt.plot(merge_x, merge_y)

plt.legend(['merge'])

plt.show()
```

## LSD sort

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('LSD sort plot', fontsize = 15)

plt.plot(lsd_x, lsd_y)

plt.legend(['LSD'])

plt.show()
```

## All sortings comparison


```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('All sortings comparison', fontsize = 15)

plt.plot(ran_x, ran_y)
plt.plot(cen_x, cen_y)
plt.plot(med_x, med_y)
plt.plot(merge_x, merge_y)
plt.plot(lsd_x, lsd_y)

plt.legend(['random', 'central', 'median', 'merge', 'LSD'])

plt.show()
```

## Выводы

LSD сортровка работает быстрее остальных, но не является универсальной, используя информацию о структуре сортируемых объектов. В частном случае сравнения чисел она хороша по времени. Среди логарифмических сортировок лидирует qsort с выбором среднего по номеру элемента, вероятно за счет отсутствия необходимости дополнительных вычислений. Merge в среднем работает чуть хуже избавленных от случайности вариантов qsort, однако лучше версии с выбором случайного опорного элемента. Квадратичные сортировки, очевидно, работают гораздо хуже логарифмтческих по времени, поэтому не были включены в итоговый график. В сравнении квадратичных сортировок удивило то, что пузырьковая в несколько раз отстает от сортровки выбором.
