# Сравнение сортировок
## Загрузка данных из файлов

```python
import matplotlib.pyplot as plt
import csv
from math import sqrt
```

```python
bubble_x = []
bubble_y = []
insertion_x = []
insertion_y = []
selection_x = []
selection_y = []

bubble_sqrt_y = []
insertion_sqrt_y = []
selection_sqrt_y = []


with open("../testing/data/bench1.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
        
        if (row[0] == "Bubble sort"):
            bubble_x.append(x)
            bubble_y.append(y)
            bubble_sqrt_y.append(sqrt(y))
        elif (row[0] == "Insertion sort"):
            insertion_x.append(x)
            insertion_y.append(y)
            insertion_sqrt_y.append(sqrt(y))
        elif (row[0] == "Selection sort"):
            selection_x.append(x)
            selection_y.append(y)
            selection_sqrt_y.append(sqrt(y))
```

```python
no_opt_x = []
no_opt_y = []
move_x = []
move_y = []
binsearch_x = []
binsearch_y = []


with open("../testing/data/bench5.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
        
        if (row[0] == "No optimization insertion sort"):
            no_opt_x.append(x)
            no_opt_y.append(y)
        elif (row[0] == "Move optimization insertion sort"):
            move_x.append(x)
            move_y.append(y)
        elif (row[0] == "Binsearch optimization insertion sort"):
            binsearch_x.append(x)
            binsearch_y.append(y)
```

```python
ran_x = []
ran_y = []
cen_x = []
cen_y = []
med_x = []
med_y = []


with open("../testing/data/bench2.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
        
        if (row[0] == "Random pivot qsort"):
            ran_x.append(x)
            ran_y.append(y)
        elif (row[0] == "Central pivot qsort"):
            cen_x.append(x)
            cen_y.append(y)
        elif (row[0] == "Median pivot qsort"):
            med_x.append(x)
            med_y.append(y)
            
```

```python
merge_x = []
merge_y = []

with open("../testing/data/bench3.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
    
        merge_x.append(x)
        merge_y.append(y)
```

```python
merge0_x = []
merge0_y = []
merge5_x = []
merge5_y = []
merge10_x = []
merge10_y = []
merge25_x = []
merge25_y = []
merge50_x = []
merge50_y = []
merge75_x = []
merge75_y = []
merge100_x = []
merge100_y = []


with open("../testing/data/bench3_cmp2.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[2])
        y = (int(row[3]) + int(row[4]) + int(row[5]) + int(row[6]) + int(row[7])) / 5
        
        if (row[0] == "Merge sort" and row[1] == "0"):
            merge0_x.append(x)
            merge0_y.append(y)
        elif (row[0] == "Merge sort" and row[1] == "5"):
            merge5_x.append(x)
            merge5_y.append(y)
        elif (row[0] == "Merge sort" and row[1] == "10"):
            merge10_x.append(x)
            merge10_y.append(y)
        elif (row[0] == "Merge sort" and row[1] == "25"):
            merge25_x.append(x)
            merge25_y.append(y)
        elif (row[0] == "Merge sort" and row[1] == "50"):
            merge50_x.append(x)
            merge50_y.append(y)
        elif (row[0] == "Merge sort" and row[1] == "75"):
            merge75_x.append(x)
            merge75_y.append(y)
        elif (row[0] == "Merge sort" and row[1] == "100"):
            merge100_x.append(x)
            merge100_y.append(y)
```

```python
lsd_x = []
lsd_y = []

with open("../testing/data/bench4.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
    
        lsd_x.append(x)
        lsd_y.append(y)
```

## Quadratic sortings

### Sortings comparison

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('Quadratic sortings comparison', fontsize = 15)

plt.plot(bubble_x, bubble_y)
plt.plot(insertion_x, insertion_y)
plt.plot(selection_x, selection_y)

plt.legend(['bubble', 'insertion', 'selection'])

plt.show()
```

## Выводы

LSD сортровка работает быстрее остальных, но не является универсальной, используя информацию о структуре сортируемых объектов. В частном случае сравнения чисел она хороша по времени. Среди логарифмических сортировок лидирует qsort с выбором среднего по номеру элемента, вероятно за счет отсутствия необходимости дополнительных вычислений. Merge в среднем работает чуть хуже избавленных от случайности вариантов qsort, однако лучше версии с выбором случайного опорного элемента. Квадратичные сортировки, очевидно, работают гораздо хуже логарифмтческих по времени, поэтому не были включены в итоговый график. В сравнении квадратичных сортировок удивило то, что пузырьковая в несколько раз отстает от сортровки выбором.
