# Сравнение сортировок
## Загрузка данных из файлов

```python
import matplotlib.pyplot as plt
import csv
from math import sqrt
```

```python
binheap_x = []
binheap_y = []

with open("../benchmarks/data/bench1.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
    
        binheap_x.append(x)
        binheap_y.append(y)
```

## Heap sort for binary heap

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('Binary Heap sort plot', fontsize = 15)

plt.plot(binheap_x, binheap_y)

plt.legend(['LSD'])

plt.show()
```
