## Загрузка данных из файлов

```python
import matplotlib.pyplot as plt
import csv
from math import sqrt
```

```python
binheap_x = []
binheap_y = []

with open("../benchmarks/data/bench1.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[1])
        y = (int(row[2]) + int(row[3]) + int(row[4]) + int(row[5]) + int(row[6])) / 5
    
        binheap_x.append(x)
        binheap_y.append(y)
```

```python
kheap5_x = []
kheap5_y = []
kheap10_x = []
kheap10_y = []
kheap25_x = []
kheap25_y = []
kheap50_x = []
kheap50_y = []
kheap100_x = []
kheap100_y = []
kheap200_x = []
kheap200_y = []
kheap300_x = []
kheap300_y = []
kheap400_x = []
kheap400_y = []
kheap500_x = []
kheap500_y = []
kheap1000_x = []
kheap1000_y = []


with open("../benchmarks/data/bench2.2.txt") as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    
    for row in data:
        x = int(row[2])
        y = (int(row[3]) + int(row[4]) + int(row[5]) + int(row[6]) + int(row[7])) / 5
        
        if (row[0] == "K-heap sort" and row[1] == "5"):
            kheap5_x.append(x)
            kheap5_y.append(y)
        elif (row[0] == "K-heap sort" and row[1] == "10"):
            kheap10_x.append(x)
            kheap10_y.append(y)
        elif (row[0] == "K-heap sort" and row[1] == "25"):
            kheap25_x.append(x)
            kheap25_y.append(y)
        elif (row[0] == "K-heap sort" and row[1] == "50"):
            kheap50_x.append(x)
            kheap50_y.append(y)
        elif (row[0] == "K-heap sort" and row[1] == "100"):
            kheap100_x.append(x)
            kheap100_y.append(y)
        elif (row[0] == "K-heap sort" and row[1] == "200"):
            kheap200_x.append(x)
            kheap200_y.append(y)
        elif (row[0] == "K-heap sort" and row[1] == "300"):
            kheap300_x.append(x)
            kheap300_y.append(y)
        elif (row[0] == "K-heap sort" and row[1] == "400"):
            kheap400_x.append(x)
            kheap400_y.append(y)
        elif (row[0] == "K-heap sort" and row[1] == "500"):
            kheap500_x.append(x)
            kheap500_y.append(y)
        elif (row[0] == "K-heap sort" and row[1] == "1000"):
            kheap1000_x.append(x)
            kheap1000_y.append(y)
```

## Heap sort for binary heap

```python
plt.xlabel('Element count', color = 'gray')
plt.ylabel('Time (ms)', color = 'gray')

plt.title('Binary Heap sort plot', fontsize = 15)

plt.plot(binheap_x, binheap_y)

plt.legend(['LSD'])

plt.show()
```
